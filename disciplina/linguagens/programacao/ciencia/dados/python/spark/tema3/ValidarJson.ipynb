```python
#python
import json, sys
path = "./clientes.json"
try:
    with open(path, "r", encoding="utf-8") as f:
        obj = json.load(f)
    print("JSON válido — tipo raiz:", type(obj).__name__)
except Exception as e:
    print("Erro ao carregar JSON:", e)
    sys.exit(1)
# opcional: reescrever formatado
with open("clientes_fixed.json", "w", encoding="utf-8") as f:
    json.dump(obj, f, ensure_ascii=False, indent=4)
```
