```python
#Array de inteiros
print("Arrays de valores inteiros")
arrayDeInteiros = [10,20,30,40,50]
print("Imprimindo todo o array:", arrayDeInteiros)
print("Tipo do array:", type(arrayDeInteiros))
print("Acessando o primeiro elemento do array:", arrayDeInteiros[0])
print("")
print("==============================================================================")
print("")

print("Arrays de valores Floats")
arrayDeFloats = [1.5,2.5,3.5,4.5,5.5]
print("Imprimindo todo o array:", arrayDeFloats)
print("Acessando o primeiro elemento do array:", arrayDeFloats[0])
print("Tipo do array:", type(arrayDeFloats))
print("")
print("==============================================================================")
print("")

print("Arrays de valores Strings")
arrayDeStrings = ["maçã","banana","laranja","uva","abacaxi"]
print("Imprimindo todo o array:", arrayDeStrings)
print("Acessando o primeiro elemento do array:", arrayDeStrings[0])
print("Tipo do array:", type(arrayDeFloats))
print("")
print("==============================================================================")
print("")
print("Arrays de valores booleans")
arrayDeBoolean = [True, False]
print("Imprimindo todo o array:", arrayDeBoolean)
print("Acessando o primeiro elemento do array:", arrayDeBoolean[0])
print("Tipo do array:", type(arrayDeBoolean))
print("")
print("==============================================================================")
print("")
print("Fim do programa")
print("Perceberam que para todos os arrays o tipo de dado é igual a list")


```

```python
print("Dicionário de dados")
print("------------------------------------------------------------------------------")
print("")
dicionarioPessoa = {
    "nome": "Roberto Carlos",
    "idade": 40,
    "altura": 1.73,
    "ligado": True
}
print("Tipo de dado do dicionário de pessoa")
print("Tipo do dicionário:", type(dicionarioPessoa))
print("")
print("==============================================================================")
print("")
print("Imprimindo todo o dicionário")
print("Dicionário:", dicionarioPessoa)
print("")
print("==============================================================================")
print("")
print("Imprimindo o nome da pessoa a partir do dicionário")
print(dicionarioPessoa['nome'])
```

```python
print("Condicionais if, else e elif")
print("------------------------------------------------------------------------------")
print("")
if(dicionarioPessoa['idade'] >= 18):
    print("A pessoa é maior de idade")
elif (dicionarioPessoa['idade'] < 18 and dicionarioPessoa['idade'] > 0):
    print("A pessoa é menor de idade")
else:
    print("Idade inválida")

if("pera" not in arrayDeStrings):
    print("A fruta pera não está no array")
else:
    print("A fruta pera está no array")
    
```

```python
print("Loop For e While")
print("------------------------------------------------------------------------------")
print("")
print("Loop For - Percorrendo o array de inteiros")
print("Usando for")
for idade in arrayDeInteiros:
    print("Idade:", idade)

print("")
print("==============================================================================")
print("")

for fruta in arrayDeStrings:
    print("Fruta:", fruta)
    
    print("")
print("==============================================================================")
print("")

for pontoflutuante in arrayDeFloats:
    print("Ponto Flutuante:", pontoflutuante)
    
print("")
print("==============================================================================")
print("")

print("Usando While")
while ligado:
    print("O sistema está ligado")
    ligado = False
```
