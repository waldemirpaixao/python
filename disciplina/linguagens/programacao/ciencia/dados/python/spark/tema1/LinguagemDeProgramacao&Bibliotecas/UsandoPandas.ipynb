```python
import json
import random
from datetime import datetime, timedelta

# Função para gerar CPF fictício
def gerar_cpf():
    return f"{random.randint(100,999)}.{random.randint(100,999)}.{random.randint(100,999)}-{random.randint(10,99)}"

# Função para gerar data de nascimento
def gerar_data_nascimento():
    hoje = datetime.now()
    data = hoje - timedelta(days=random.randint(7300, 25550))  # Entre 20 e 70 anos
    return data.strftime("%Y-%m-%d")

# Lista de clientes
clientes = []
for i in range(100):
    cliente = {
        "id": i + 1,
        "nome": f"Cliente_{i+1}",
        "email": f"cliente{i+1}@email.com",
        "cpf": gerar_cpf(),
        "data_nascimento": gerar_data_nascimento(),
        "telefone": f"(11) 9{random.randint(1000,9999)}-{random.randint(1000,9999)}",
        "endereco": {
            "rua": f"Rua {random.randint(1,100)}",
            "numero": random.randint(1,1000),
            "bairro": random.choice(["Centro", "Jardim", "Vila Nova", "Boa Vista", "Santa Cruz"]),
            "cidade": random.choice(["São Paulo", "Rio de Janeiro", "Belo Horizonte", "Curitiba", "Porto Alegre"]),
            "estado": random.choice(["SP", "RJ", "MG", "PR", "RS"]),
            "cep": f"{random.randint(10000,99999)}-{random.randint(100,999)}"
        },
        "historico_compras": {
            "total_compras": random.randint(1,20),
            "valor_total_gasto": round(random.uniform(100,10000), 2),
            "ultima_compra": (datetime.now() - timedelta(days=random.randint(0,365))).strftime("%Y-%m-%d")
        },
        "preferencias": {
            "categorias_favoritas": random.sample([
                "Hardware",
                "Periféricos",
                "Notebooks",
                "Monitores",
                "Acessórios"
            ], k=random.randint(1,3)),
            "forma_pagamento_preferida": random.choice(["Crédito", "Débito", "PIX", "Boleto"])
        },
        "ativo": random.choice([True, False])
    }
    clientes.append(cliente)

# Salvar como JSON
with open('clientes.json', 'w', encoding='utf-8') as f:
    json.dump({"clientes": clientes}, f, ensure_ascii=False, indent=4)

print("Arquivo clientes.json criado com sucesso!")
```

```python
# Gerar JSON Lines (cada cliente em uma linha)
import json
from datetime import datetime

clientes = []
for i in range(1, 11):  # 10 clientes como exemplo
    cliente = {
        "id": i,
        "nome": f"Cliente_{i}",
        "email": f"cliente{i}@email.com",
        "cidade": "São Paulo" if i % 3 == 0 else "Rio de Janeiro",
        "categoria_favorita": "Hardware",
        "ultima_compra": "2023-11-06",
        "total_compras": i * 100,
        "ativo": True if i % 2 == 0 else False
    }
    clientes.append(cliente)

# Salvar como JSON Lines (um objeto por linha)
with open('clientes.jsonl', 'w', encoding='utf-8') as f:
    for cliente in clientes:
        f.write(json.dumps(cliente, ensure_ascii=False) + '\n')

print("Arquivo clientes.jsonl criado com sucesso!")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Dados para geração de registros
produtos = {
    'Hardware': ['Processador i5', 'Processador i7', 'Placa de Vídeo RTX 3060', 'SSD 500GB', 'Memória RAM 8GB'],
    'Periféricos': ['Mouse Gamer', 'Teclado Mecânico', 'Headset 7.1', 'Monitor 24"', 'Webcam HD'],
    'Acessórios': ['Mousepad', 'Suporte Monitor', 'Hub USB', 'Cabo HDMI', 'Adaptador HDMI']
}

precos = {
    'Processador i5': 1200.00,
    'Processador i7': 2200.00,
    'Placa de Vídeo RTX 3060': 2500.00,
    'SSD 500GB': 350.00,
    'Memória RAM 8GB': 200.00,
    'Mouse Gamer': 150.00,
    'Teclado Mecânico': 300.00,
    'Headset 7.1': 250.00,
    'Monitor 24"': 900.00,
    'Webcam HD': 180.00,
    'Mousepad': 50.00,
    'Suporte Monitor': 120.00,
    'Hub USB': 80.00,
    'Cabo HDMI': 45.00,
    'Adaptador HDMI': 35.00
}

# Criar registros de vendas
n_registros = 1000
dados = {
    'ID_Venda': range(1, n_registros + 1),
    'Data': [(datetime.now() - timedelta(days=int(x))).strftime('%Y-%m-%d') 
             for x in sorted(np.random.randint(0, 365, n_registros))],
    'Produto': np.random.choice(list(precos.keys()), n_registros),
    'Categoria': '',  # será preenchido depois
    'Quantidade': np.random.randint(1, 6, n_registros),
    'Preco_Unitario': 0,  # será preenchido depois
    'Valor_Total': 0,  # será preenchido depois
    'Forma_Pagamento': np.random.choice(['Crédito', 'Débito', 'PIX', 'Boleto'], n_registros),
    'Vendedor': np.random.choice(['João Silva', 'Maria Santos', 'Pedro Oliveira', 'Ana Costa'], n_registros),
    'ID_Cliente': np.random.randint(1000, 9999, n_registros)
}

# Criar DataFrame
df = pd.DataFrame(dados)

# Preencher categoria e preços
for categoria, items in produtos.items():
    for produto in items:
        mask = df['Produto'] == produto
        df.loc[mask, 'Categoria'] = categoria
        df.loc[mask, 'Preco_Unitario'] = precos[produto]

# Calcular valor total
df['Valor_Total'] = df['Quantidade'] * df['Preco_Unitario']

# Ordenar por data
df = df.sort_values('Data')

# Salvar como CSV
df.to_csv('vendas_ti.csv', index=False)

print("Arquivo vendas_ti.csv criado com sucesso!")
print("\nPrimeiros registros:")
print(df.head())
print("\nEstatísticas:")
print(df.describe())
```
