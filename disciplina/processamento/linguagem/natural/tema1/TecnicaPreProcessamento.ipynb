```python

import os
import locale
from nltk.corpus import PlaintextCorpusReader


def criar_corpus_machado_simples():
    """Cria um corpus com trechos das obras do Machado"""
    locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
    
    #corpus_dir = "./machado_corpus" 
    corpus_dir = "./nltk_data/corpora/machado"
    os.makedirs(corpus_dir, exist_ok=True)
    
    # Trechos das principais obras
    obras = {
        "dom_casmurro.txt": """
        Alguns anos havia eu entrado no foro. Trabalhava, conseguia,
        e ia ganhando algum dinheiro. N√£o me dei √† pol√≠tica, n√£o
        fui √† C√¢mara, n√£o fui ao Pa√ßo. As minhas rela√ß√µes de foro
        deram-me algumas da sociedade, e estas outras, e entrei a
        ser convidado para jantar, para bailes, para soir√©es.
        """,
        
        "memorias_postumas.txt": """
        Ao verme que primeiro roeu as frias carnes do meu cad√°ver
        dedico como saudosa lembran√ßa estas mem√≥rias p√≥stumas.
        """,
        
        "quincas_borba.txt": """
        Ao chegar √† Rua da Harmonia, Rubi√£o fitou os olhos no solar,
        e perguntou a si mesmo se era ali que morava. Era um solar
        antigo, mas bem conservado.
        """
    }
    
    # Salvar arquivos
    for nome, texto in obras.items():
        with open(os.path.join(corpus_dir, nome), 'w', encoding='utf-8') as f:
            f.write(texto)
    
    return PlaintextCorpusReader(corpus_dir, '.*\.txt')

# Usar o corpus
machado = criar_corpus_machado_simples()
print("‚úÖ Corpus Machado criado!")
# üî• AGORA VAMOS USAR COMO O CORPUS ORIGINAL:
print("=" * 50)
print("üìö COMO USAR O CORPUS MACHADO CRIADO")
print("=" * 50)

# 1. Listar arquivos dispon√≠veis
print("1. üìÅ ARQUIVOS DISPON√çVEIS:")
print(machado.fileids())
print()

# 2. Acessar palavras
# print("2. üî§ PALAVRAS (todas as obras):")
# palavras_todas = machado.words()
# print(f"Total de palavras: {len(palavras_todas)}")
# print(f"Primeiras 15 palavras: {palavras_todas[:15]}")
# print()

# 3. Palavras de um arquivo espec√≠fico
print("3. üìÑ PALAVRAS DE UM ARQUIVO ESPEC√çFICO:")
palavras_dom_casmurro = machado.words('dom_casmurro.txt')
print(f"Dom Casmurro - Primeiras 10 palavras: {palavras_dom_casmurro[:10]}")
print()

# 4. Senten√ßas
# print("4. üìù SENTEN√áAS:")
# sentencas_todas = machado.sents()
# print(f"Total de senten√ßas: {len(sentencas_todas)}")
# print(f"Primeiras 3 senten√ßas:")
# for i, sentenca in enumerate(sentencas_todas[:3], 1):
#     print(f"  {i}. {' '.join(sentenca)}")
# print()

# 5. Texto bruto
print("5. üìñ TEXTO BRUTO:")
texto_bruto = machado.raw('memorias_postumas.txt')
print(f"Texto completo (primeiros 100 chars): {texto_bruto[:100]}...")
print()
```
