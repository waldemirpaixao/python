```python
import cv2 as cv
import matplotlib.pyplot as plt


img1 = cv.imread('../imagens/audi1.jpeg',cv.IMREAD_GRAYSCALE)
# img2 = cv.imread('../imagens/audi2.jpeg', cv.IMREAD_GRAYSCALE)
# img3 = cv.imread('../imagens/audi3.jpeg', cv.IMREAD_GRAYSCALE)
# img4 = cv.imread('../imagens/audi4.jpeg', cv.IMREAD_GRAYSCALE)
# img5 = cv.imread('../imagens/audi6.jpeg', cv.IMREAD_GRAYSCALE)

img = cv.cvtColor(img1, cv.COLOR_RGB2BGR)
plt.imshow(img1)


orb = cv.ORB_create()
kp = orb.detect(img1,None)

kp, des = orb.compute(img1, kp)

img2 = cv.drawKeypoints(img1, kp, None, color=(0,255,0), flags=0)
plt.imshow(img2), plt.show()
```
