```python
# 1. Instalação e Importação
# Generated by DeepSeek

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Para exibir imagens no notebook (opcional)
%matplotlib inline


#2. Carregar e Pré-processar a Imagem
# Carregar imagem em escala de cinza
img = cv2.imread('../imagens/audi.png', cv2.IMREAD_GRAYSCALE)

# Ou converter colorida para cinza
# img_color = cv2.imread('sua_imagem.jpg')
# img = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

# Verificar se a imagem foi carregada
if img is None:
    print("Erro: Não foi possível carregar a imagem")
else:
    print(f"Dimensões: {img.shape}")
    
    
#3. Limiarização Simples (Binary Threshold)
# Aplicar limiarização simples
valor_limiar = 127
valor_maximo = 255

# retval, img_limiarizada = cv2.threshold(imagem, limiar, valor_max, tipo)
ret, thresh_binary = cv2.threshold(img, valor_limiar, valor_maximo, cv2.THRESH_BINARY)

print(f"Limiar usado: {ret}")

#4. Tipos de Limiarização
# Diferentes tipos de limiarização
_, thresh_binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
_, thresh_binary_inv = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
_, thresh_trunc = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
_, thresh_tozero = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
_, thresh_tozero_inv = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)


#5. Limiarização Adaptativa
# Limiarização adaptativa - útil para imagens com iluminação irregular
thresh_adapt_mean = cv2.adaptiveThreshold(
    img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
    cv2.THRESH_BINARY, 11, 2
)

thresh_adapt_gaussian = cv2.adaptiveThreshold(
    img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
    cv2.THRESH_BINARY, 11, 2
    
)

#6. Limiarização de Otsu
# Otsu's method - encontra automaticamente o melhor limiar
ret2, thresh_otsu = cv2.threshold(
    img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
)
print(f"Limiar Otsu: {ret2}")


```
