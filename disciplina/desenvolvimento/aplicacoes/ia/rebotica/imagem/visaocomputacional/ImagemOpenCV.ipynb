```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
print("Versão do OpenCV",cv2.__version__)
print("Versão do Matplotlib", matplotlib.__version__)

imagem = cv2.imread("./imagens/audi.png")
imagem - cv2.cvtColor(imagem, cv2.COLOR_RGB2BGR)

print("Exibindo a imagem")
plt.imshow(imagem)

```

```python
print("Histograma da imagem")
plt.hist(imagem.ravel())


```

```python
imgRavel = imagem.ravel()
print(imgRavel)
```

```python
print("Limpando a tela de plot")
plt.clf()
```
